# Image *20.jpg

```python
from __future__ import print_function
from imutils.object_detection import non_max_suppression
from imutils import paths
import numpy as np
import argparse
import imutils
import cv2

img_path = 'IMG_20170814_163020.jpg'

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())


image = cv2.imread(img_path)
image = imutils.resize(image, width=min(600, image.shape[1]))
orig = image.copy()
O_image = imutils.resize(image, width=min(600, image.shape[1]))

(rects, weights) = hog.detectMultiScale(image, winStride=(3, 3),
    padding=(0, 256), scale = 1.04, hitThreshold = 0.095, group_threshold=2)
for (x, y, w, h) in rects:
    cv2.rectangle(orig, (x, y), (x + w, y + h), (0, 0, 255), 2)
rects = np.array([[x, y, x + w, y + h] for (x, y, w, h) in rects])
pick = non_max_suppression(rects, probs=None, overlapThresh=0.85)
for (xA, yA, xB, yB) in pick:
    cv2.rectangle(image, (xA, yA), (xB, yB), (0, 255, 0), 2)

print("[INFO] {}: {} original boxes, {} after suppression".format(
    img_path, len(rects), len(pick)))
cv2.startWindowThread()
cv2.imshow("Before NMS", O_image)
cv2.imshow("After NMS", image)
cv2.waitKey(1)
cv2.destroyAllWindows()
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
cv2.waitKey(1)
```
